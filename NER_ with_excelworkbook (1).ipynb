```python
# write text file with the name myfile.txt
with open("myfile.txt", 'w') as f: 
    for key, value in df.items(): 
        f.write('%s:%s\n' % (key, value))
```

```python
with open("myfile.txt") as f:
    contents = f.read()
    print(contents)
```

```python
contents
```

```python
def print_entities(pipeline, text):
    
    # Create a document 
    document = pipeline(text)
    
    # Entity text & label extraction
    for entity in document.ents:
        print(entity.text + '->', entity.label_)
        
        
def visualize_entities(pipeline, text):
    
    # Create a document 
    document = pipeline(text)
        
    # Show entities in pretty manner
    displacy.render(document, jupyter=True, style='ent')
```

```python
import spacy
from spacy import displacy

# Load English large model
nlp_sm = spacy.load("en_core_web_lg")
```

```python
print_entities(nlp_sm, contents)
```

```python
visualize_entities(nlp_sm,contents)
```

# Let’s see how the roBERTa model behaves on the same texts.



```python
# Load the spacy transformer (roberta-base) model
roberta_nlp = spacy.load("en_core_web_trf")
```

```python
print_entities(roberta_nlp,contents)
```

```python
visualize_entities(roberta_nlp, contents)
```
